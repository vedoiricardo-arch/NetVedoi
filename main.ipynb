**EXERCICIO LISTA + INPUT SE TEM NOME NA LISTA E A POSIÇÃO**

```python
nomes = ["nerzilus", "ricardo", "vedoi", "joao", "carlos"]
print(nomes)

nome = input("informe o nome da lista")

if nome in nomes:
  posicao = nomes.index(nome) + 1
  print(f"'{nome}' encontrado na posição: {posicao}")

else:
    print(f"'{nome}' não encontrado na lista.")
```

```python
import random # Importa o módulo random para gerar elementos aleatórios

# Listas de partes de nomes para criar nomes aleatórios
primeiros_nomes = ["João", "Maria", "Pedro", "Ana", "Carlos", "Mariana", "Fernando", "Patrícia", "Rafael", "Juliana"]
sobrenomes = ["Silva", "Santos", "Oliveira", "Souza", "Pereira", "Almeida", "Costa", "Rodrigues", "Martins", "Gomes"]

nomes_aleatorios = [] # Inicializa uma lista vazia para armazenar os nomes gerados

# Loop para gerar 100 nomes aleatórios
for i in range(100):
    # Escolhe um primeiro nome aleatório da lista 'primeiros_nomes'
    primeiro = random.choice(primeiros_nomes)
    # Escolhe um sobrenome aleatório da lista 'sobrenomes'
    sobrenome = random.choice(sobrenomes)
    # Combina o primeiro nome e o sobrenome para formar um nome completo
    nome_completo = f"{primeiro} {sobrenome}"
    # Adiciona o nome completo à lista 'nomes_aleatorios'
    nomes_aleatorios.append(nome_completo)

# Imprime todos os nomes aleatórios gerados
print("Lista de 100 nomes aleatórios:")
for nome in nomes_aleatorios:
    print(nome)
```
